# Stack Overflow Link/ID Uniqueness

```python
from urllib.parse import urlparse
```

```python
# Insert column contnaining Stack Overflow ID
df.insert(df.columns.get_loc('so_link')+1, 'so_id', 0)
df['so_id'] = df['so_id'].astype(int)
df['so_link'] = df['so_link'].astype(str)
```

```python
# Add column containing only the ID's of the Stack Overflow posts chosen

# o = urlparse('https://stackoverflow.com/questions/40130482/indexerror-pop-index-out-of-range-with-a-for-loop')
# o.path.split('/')[2] # This is the ID

def label_link_id(row):
    if row['so_link'].count('\n'):
        return 0
    if row['so_link'] == 'nan':
        # empty, NaN
        return 0
    return int(urlparse(row['so_link']).path.split('/')[2])

#df.apply (lambda row: label_link_id(row), axis=1)
df['so_id'] = df.apply (lambda row: label_link_id(row), axis=1)
```

```python
# Group data by stimulus ID
id_groups = df.groupby('stimulus_id')

# Example (for reference)
group = id_groups.get_group(0)
print("Stimulus ID 0, Stack Overflow ID 27705523 has",group[group.so_id == 27705523].so_id.count(),"occurrences") # index by stimulus ID, so ID
```

```python
# Insert column indicating whether or not link is unique within this stimulus group
df.insert(df.columns.get_loc('so_link')+1, 'unique_link', True)
```

```python
# Populate 'unique_link' column based on whether this link is unique
# among all participant responses
id_groups = df.groupby('stimulus_id')

def label_unique_link(row):
    group = id_groups.get_group(row['stimulus_id'])
    if row['so_id'] == 0:
        return True
    if group[group.so_id == row['so_id']].so_id.count() > 1:
        return False # Not unique
    return True

#df.apply (lambda row: label_link_id(row), axis=1)
df['unique_link'] = df.apply (lambda row: label_unique_link(row), axis=1)
df
```

## Chi-Square Test

```python
# Perform contingency test for experience level vs. link uniqueness (for each participant response)
chi_test(df, 'xp_level', [df.unique_link])
```

```python
def rank_corr(df, col: str):
    """REQUIRES: col parameter MUST be a boolean column in df"""
    # Heavily referenced:
    # https://medium.com/analytics-vidhya/testing-a-difference-in-population-proportions-in-python-89d57a06254
    # https://machinelearningmastery.com/how-to-calculate-nonparametric-rank-correlation-in-python/

    #expert_relevant = df[df[col]]
    expert_relevant = df[df['xp_level'] == 'Expert']
    expert_grouped = expert_relevant.groupby('error_type')

    #novice_relevant = df[df[col]]
    novice_relevant = df[df['xp_level'] != 'Expert']
    novice_grouped = novice_relevant.groupby('error_type')

    # Get proportion of rows with TRUE col / total
    expert_proportions = pd.crosstab(expert_relevant[col],expert_relevant['error_type']).apply(lambda r: r / r.sum(), axis = 0) #Contingency Table
    novice_proportions = pd.crosstab(novice_relevant[col],novice_relevant['error_type']).apply(lambda r: r / r.sum(), axis = 0)
    display(pd.crosstab(df[col],df['error_type']).apply(lambda r: r / r.sum(), axis = 0))

    #coef, p = spearmanr(expert_ranks, novice_ranks)
    expert_ranks = expert_proportions.T.reset_index().sort_values(by=True,ascending = False).reset_index()
    novice_ranks = novice_proportions.T.reset_index().sort_values(by=True,ascending = False).reset_index()
    # At this point, column 'index' in expert_ranks and novice_ranks will
    # hold the original number (1,2,...,6) assocated with each error type
    print(color.bold('expert'))
    display(expert_ranks)
    print(color.bold('novice'))
    display(novice_ranks)
    # calculate spearman's correlation
    coef, p = spearmanr(expert_ranks['index'], novice_ranks['index'])
    print('Spearmans correlation coefficient: %.3f' % coef)
    # interpret the significance
    alpha = 0.05
    if p > alpha:
        print('Expert and novice ' + col + ' proportions are uncorrelated (fail to reject H0) p=%.3f' % p)
    else:
        print('Expert and novice ' + col + ' proportions are correlated (reject H0) p=%.3f' % p)
    
```
